```python
PATH_GOOGLE_MOBILITY_appel = "applemobilitytrends-2020-05-05.csv"
df_appel = pd.read_csv(PATH_GOOGLE_MOBILITY_appel,error_bad_lines=False)
# df_appel=df_appel[df_appel['geo_type'] == "country/region"]
df_appel
```

```python
df_appel= df_appel.drop(columns=["geo_type", "alternative_name"])
df_appel = df_appel.melt(id_vars= ["region", "transportation_type"])
df_appel
```

```python
df_appel.head(4)
```

```python
df_appel_europe = df_appel[df_appel.region.isin(pays_europe)]
df_appel_europe =  df_appel_europe.rename(columns={"region": "Country"})
df_appel_europe
```

```python
dd=df_appel_europe.groupby(['Country',"variable","transportation_type" ]).sum()
ff=dd.reset_index()
appel_europe= ff.rename(columns={"variable": "date", "value": "total_mobilite"})
appel_europe['date'] =pd.to_datetime(appel_europe.date)

appel_europe = appel_europe.sort_values(by = "date")
appel_europe
```

```python
appel_europe.to_csv("appel_europe.csv")
```
